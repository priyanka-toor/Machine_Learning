```python
import pandas as pd
data=pd.read_excel('data.xlsx')
data.info()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

X = data.drop(columns=['Age'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=1)  
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data['Age'], cmap='viridis')  # Use the target column for coloring points
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Plot')
plt.colorbar(label='Target')  # Add a color bar for reference

# Show the plot
plt.show()



```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Perform PCA with two components
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=data['Age'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Plot')
plt.colorbar(label='Age')  # Color points based on 'Age'

# Show the plot
plt.show()
```
